```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
data=pd.read_excel('company descriptions.xlsx')
```

```python
data
```

```python
s=set(stopwords.words('english'))
```

```python
desc=data.company_short_description
```

```python
import re
def remove(sent):
    sent=sent.lower()
    sent=re.sub(r'[^a-zA-Z_\s]+', ' ', sent)
    total_sent=[]
    for i in sent.split():
        if i not in s:
            total_sent.append(i)
    return " ".join(total_sent)
```

```python
value=remove(desc[0])
```

```python
sec_data=pd.read_excel('Industry Segments - Top 10 Keywords.xlsx')
```

```python
kw_dict={}
for i in sec_data.values[:26]:
    kw_dict[i[0]]=i[1].split(',')
```

```python
kw_dict
```

```python
def get_industry(value):
    maximum=0
    industry=None
    for k,v in kw_dict.items():
        v=[i.lower().strip() for i in v]
        keyword_match=len(set(value.split()).intersection(set(v)))
        if keyword_match>maximum:
            industry=k
            maximum=keyword_match
    return industry
```

```python
get_industry(value)
```

```python
c=[]
for i in desc.values:
    clean=remove(i)
    print(get_industry(clean))
    c.append(get_industry(clean))
```

```python
data['industry_label']=c
data.to_csv('industry_label_matching.csv')
data
```

#### Problem Statement :- 
Write a code to match names from list1 (provided in the file: name matching -list1.xlsx) to list2 (name matching â€“ list2.xlsx). The names in list1 have multiple variations of a name and map to exactly one name in list2. Also, provide an accuracy measure of the names matched. 

```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import warnings
warnings.filterwarnings("ignore")
```

```python
data1=pd.read_excel('name matching - list1.xlsx',names=['list1'])
data2=pd.read_excel('name matching task - list2.xlsx',names=['list2'])
```

```python
data1['list1']=data1['list1'].str.replace(r"\([^()]*\)","")
```

```python
df1=data1['list1'].tolist()
df2=data2['list2'].tolist()
```

#### Fuzzywuzzy 
In computer science, fuzzy string matching is the technique of finding strings that match a pattern approximately (rather than exactly). In another word, fuzzy string matching is a type of search that will find matches even when users misspell words or enter only partial words for the search. It is also known as approximate string matching.

```python
score=[]
for j in range(0,len(df2)):
    score.append(process.extractOne(df1[j],df2,scorer=fuzz.token_set_ratio))
    print(process.extractOne(df1[j],df2,scorer=fuzz.token_set_ratio))
```

```python
q=data1
match=[i[0] for i in score ]
similarity_score=[i[1] for i in score ]

q['match']=match
q['similarity_score']=similarity_score
q.to_csv('matching.csv')
```
